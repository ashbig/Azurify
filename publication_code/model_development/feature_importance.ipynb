```python
from catboost import CatBoostClassifier
compM = CatBoostClassifier()
compM.load_model('azurify.json', format='json')
```

```python
df = compM.get_feature_importance(prettified=True)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
plt.rcParams['font.family'] = 'Arial'
plt.rcParams["figure.figsize"] = (8,6)
plt.rcParams.update({'font.size': 16})
ax = fig.add_axes([0,0,1,1])
plt.style.use('seaborn-white')
plt.rcParams['font.size'] = 16
plt.rcParams["font.family"] = "Arial"
ax.bar(df['Feature Id'], df['Importances'],  color='#011F5b')
labels=['Effect', 'Allele Freq.', 'Gene', 'Protein Change', 'ClinVar Class', 'In silico (MVP)', 'Protein Domain', 'COSMIC Count', 'Exon #', 'gnomAD Freq.', 'PubMed #', 'KEGG', 'gnomAD Count', 'CIVic Count', 'Drug Target']
ax.set_xticklabels(labels, rotation = 45, ha="right")
plt.xlabel('\nFeature')
plt.ylabel('% Class Contribution\n')
plt.rcParams["figure.figsize"] = (10,6)
plt.savefig('FeatureIMP.pdf', format="pdf", bbox_inches='tight', dpi=600)
plt.show()
```
